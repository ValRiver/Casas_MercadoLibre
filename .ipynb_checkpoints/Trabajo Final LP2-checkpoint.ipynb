<p style="font-size:25px;color:skyblue";>Descripci√≥n del proyecto:</p>  

El presente trabajo consiste en recopilar los datos principalmente de los precios, √°rea ($m^{2}$), ubicaci√≥n, n√∫mero de habitaciones y otros atributos de las casas que est√°n en venta en la ciudad de Lima, esta informaci√≥n se obtendr√° haciendo un web scraping de la p√°gina web Mercado Libre Per√∫, posteriormente se crear√° un primer data frame solo con los datos de la primera p√°gina webscrapeada. La b√∫squeda de las casas en Lima registran un total de cuatro p√°ginas, para ello se crear√° una iteraci√≥n haciendo de nuevo un web scraping para as√≠ obtener los cambios de p√°gina y utilizando los c√≥digos ya obtenidos en un principio, finalmente se crea un archivo excel, para guardar el trabajo realizado.   
  
<p style="font-size:25px;color:skyblue";>Limitaciones:</p>  

- En la primera p√°gina de resultados de la busqueda "Casas Lima" aparecen 48 coincidencias, al momento de obtener el n√∫mero de habitaciones de todas estas se registran s√≥lo 47 de ellas verificamos que en una descripci√≥n de casa no se registro esa informaci√≥n y por ende no contiene la etiqueta requerida para rastrear su informaci√≥n. M√°s adelante observamos tambien que la p√°gina tres tiene el mismo defecto,entonces optamos por no utilizar el n√∫mero de habitaciones para el data frame, ya que al no encontrar este dato, en una de las casas, este provocar√≠a un desorden en las filas.  
- Encontramos otra limitaci√≥n, al momento de observar la p√°gina, vemos la variaci√≥n de estos y en el tipo de moneda que se encuentra (soles o d√≥lares), luego para el momento de evaluar los precios se deber√≠a tener en cuenta en el tipo de moneda que se ha usado, de igual forma optamos por webscarpear la cantidad registrada y el tipo de moneda por separado. 

Nota: Verificar siempre el # de datos obtenidos por p√°gina.

```python
#Librerias a usar:
import requests  #hacer peticiones http
from bs4 import BeautifulSoup  #Extraer informaci√≥n en html
from lxml import etree #la libreria lxml nos permite un mejor manejo de la informaci√≥n en html, 
import pandas as pd #Creaci√≥n de data frames
```

```python
#Se realiza la solicitud de importaci√≥n:
r=requests.get("https://listado.mercadolibre.com.pe/casas-lima") 
r 
```

```python
# Extracci√≥n de informaci√≥n en formato HTML con la libreria BeautifulSoup: 
soup=BeautifulSoup(r.text,"html.parser") 
# lectura con r.text ya que se tendr√° el texto completo con tildes y √±s sin errores de ortograf√≠a
```

```python
soup
```

<p style="font-size:22px;color:green";>Extracci√≥n de t√≠tulos</p>

```python
#Con "find_all" buscamos la l√≠nea que contiene los t√≠tulos (obtenida de la previa inspecci√≥n hecha en la p√°gina web)
titulos=soup.find_all("h2",attrs={"class":"ui-search-item__title ui-search-item__group__element"}) 
```

```python
#Limpieza: Se requiere s√≥lo el texto de los t√≠tulos:
titulos=[i.text for i in titulos]
print(titulos)
```

<p style="font-size:22px;color:green";>Extracci√≥n del tipo de moneda</p>

```python
#Buscamos el tipo de moneda con la funci√≥n find_all (damos la etiquetas obtenidas previa inspecci√≥n en la p√°gina):
moneda=soup.find_all("span",attrs= {"class":"price-tag-symbol"})
```

```python
# Limpieza:
moneda=[i.text for i in moneda ]
print(moneda)
```

<p style="font-size:22px;color:green";>Extracci√≥n de los precios</p>

```python
#Buscamos los precios con la funci√≥n find_all (damos la etiquetas obtenidas previa inspecci√≥n en la p√°gina):
precios=soup.find_all("span",attrs= {"class":"price-tag-fraction"})
```

```python
# Limpieza:
precios=[i.text for i in precios]
print(precios)
```

<p style="font-size:22px;color:green";>Extracci√≥n del √°rea en $m^{2}$</p>

```python
#Usamos el xpath para encontrar un elemento en particular de la p√°gina:
dom=etree.HTML(str(soup))
area=dom.xpath('//ul[@class="ui-search-card-attributes ui-search-item__group__element"]//li[1]')
```

```python
#Limpieza:
area=[i.text for i in area]
print(area)
# La extracci√≥n del √°rea de las casas est√° en conjunto con su valor en $m^{2}$ m√°s dos categorias (cubierta o total),
# decidimos dejarlo de esa manera.
```

<p style="font-size:22px;color:green";>Extracci√≥n de la ubicaci√≥n / direcci√≥n</p>

```python
#Buscamos las direcciones de las casas mediante inspecci√≥n a la p√°gina, luego usamos la funci√≥n find_all: 
ubicacion=soup.find_all("span",attrs={"class":"ui-search-item__group__element ui-search-item__location"})
```

```python
#Limpieza
ubicacion=[i.text for i in ubicacion]
print(ubicacion)
```

<p style="font-size:22px;color:green";>Extracci√≥n del n√∫mero de habitaciones</p>

```python
#Usamos el xpath:
habitaciones=dom.xpath('//ul[@class="ui-search-card-attributes ui-search-item__group__element"]//li[2]')
```

```python
#Limpieza:
habitaciones=[i.text for i in habitaciones]
print(habitaciones)
```

<p style="font-size:22px;color:green";>Extracci√≥n de URLs</p>

```python
#Se extrae de manera similar a los t√≠tulos, modificando etiquetas y atributos:
urls=soup.find_all("a",attrs={"class":"ui-search-result__content ui-search-link"})
```

```python
# Limpieza para los urls, se quiere obtener s√≥lo los links contenidos en "href"
urls=[i.get("href") for i in urls]
print(urls)
```

```python
#Requerimos el mismo n√∫mero de observaciones para lograr crear el data frame posterior
print(len(titulos))
print(len(moneda))
print(len(precios))
print(len(area))
print(len(ubicacion))
print(len(habitaciones))
print(len(urls))
# Notamos que el npumero de habitaciones es menor por lo cual decidimos no incluirla en el data frame.
```

```python
#creando una lista donde se guardaran todas las paginas de cada informaci√≥n como titulos,precios,areas,ubicacion y url's
lista_titulos = []
lista_precios = []
lista_areas =   []
lista_ubicacion = []
lista_url =   []
lista_soles_dolares = []
```

```python
#creamos el bucle para recorrer todas las paginas de mercado libre 
from requests.api import request
siguiente = 'https://listado.mercadolibre.com.pe/casas-lima'  #en principio lee el html de la pagina 1 del objeto'siguiente'
while True: 
    r= requests.get(siguiente)                  #con requests se obtendra el html de la pagina y asi sucesivamente recorrer√° hasta la √∫ltima pagina 
    if r.status_code==200:                       #validando el status a 200
        soup = BeautifulSoup(r.text,'html.parser')
        #####Titulos
        titulos=soup.find_all("h2",attrs={"class":"ui-search-item__title ui-search-item__group__element"}) 
        titulos=[i.text for i in titulos]
        lista_titulos.extend(titulos)              # extendiendo la lista sobre la informacion de los titulos

        #####precios
        precios=soup.find_all("span",attrs= {"class":"price-tag-fraction"})
        precios=[i.text for i in precios]
        lista_precios.extend(precios)             # extendiendo la lista sobre la informacion de los precios 


        #####areas
        dom=etree.HTML(str(soup))
        area=dom.xpath('//ul[@class="ui-search-card-attributes ui-search-item__group__element"]//li[1]')
        area=[i.text for i in area]
        lista_areas.extend(area)                  # extendiendo la lista sobre la informacion de las areas


        #####ubicacion
        ubicacion=soup.find_all("span",attrs={"class":"ui-search-item__group__element ui-search-item__location"})
        ubicacion=[i.text for i in ubicacion]
        lista_ubicacion.extend(ubicacion)        # extendiendo la lista sobre la informacion de las ubicaciones 


        ####url:
        urls=soup.find_all("a",attrs={"class":"ui-search-result__content ui-search-link"})
        urls=[i.get("href") for i in urls]
        lista_url.extend(urls)                  # extendiendo la lista sobre la informacion de los url

        ####soles_dolares:
        moneda=soup.find_all("span",attrs= {"class":"price-tag-symbol"})
        moneda=[i.text for i in moneda ]
        lista_soles_dolares.extend(moneda)

        ini=soup.find("span",attrs={"class":"andes-pagination__link"}).text
        ini=int(ini)                        #volver a validar la pagina inicial para el segundo recorrido del html en el bucle 

        can=soup.find("li",attrs={"class":"andes-pagination__page-count"})
        can=int(can.text.split(" ")[1])    #volver a validar la cantidad de paginas 

    


    else:
        break
    print(ini,can)                              

    if ini==can:  
        break             #break si ambos coinciden con la ultima pagina
    

    siguiente = dom.xpath('//div[@class="ui-search-pagination"]/ul/li[contains(@class,"--next")]/a')[0].get('href')
```

```python
#longitudes de las listas creadas (ambas deberian ser del mismo tama√±o)
print(len(lista_titulos))
print(len(lista_precios)) 
print(len(lista_areas))
print(len(lista_ubicacion))
print(len(lista_url))
print(len(lista_soles_dolares))
```

Data frame de todas las paginas extraidas (1-4)

```python
df_final=pd.DataFrame({"T√≠tulos":lista_titulos, "Monedas":lista_soles_dolares,"Precios":lista_precios,"√Årea":lista_areas,"URL":lista_url,"Ubicaciones":lista_ubicacion})
df_final
```

```python
data=pd.DataFrame(df_final,columns=["T√≠tulos","Monedas", "Precios","√Årea","URL","Ubicaciones"])
data
```
