<p style="font-size:25px;color:skyblue";>Descripción del proyecto:</p>  

El presente trabajo consiste en recopilar los datos principalmente de los precios, área ($m^{2}$), ubicación, número de habitaciones y otros atributos de las casas que están en venta en la ciudad de Lima, esta información se obtendrá haciendo un web scraping de la página web Mercado Libre Perú, posteriormente se creará un primer data frame solo con los datos de la primera página webscrapeada. La búsqueda de las casas en Lima registran un total de cuatro páginas, para ello se creará una iteración haciendo de nuevo un web scraping para así obtener los cambios de página y utilizando los códigos ya obtenidos en un principio, finalmente se crea un archivo excel, para guardar el trabajo realizado.   
  
<p style="font-size:25px;color:skyblue";>Limitaciones:</p>  

- En la primera página de resultados de la busqueda "Casas Lima" aparecen 48 coincidencias, al momento de obtener el número de habitaciones de todas estas se registran sólo 47 de ellas verificamos que en una descripción de casa no se registro esa información y por ende no contiene la etiqueta requerida para rastrear su información. Más adelante observamos tambien que la página tres tiene el mismo defecto,entonces optamos por no utilizar el número de habitaciones para el data frame, ya que al no encontrar este dato, en una de las casas, este provocaría un desorden en las filas.  
- Encontramos otra limitación, al momento de observar la página, vemos la variación de estos y en el tipo de moneda que se encuentra (soles o dólares), luego para el momento de evaluar los precios se debería tener en cuenta en el tipo de moneda que se ha usado, de igual forma optamos por webscarpear la cantidad registrada y el tipo de moneda por separado. 

Nota: Verificar siempre el # de datos obtenidos por página.

```python
#Librerias a usar:
import requests  #hacer peticiones http
from bs4 import BeautifulSoup  #Extraer información en html
from lxml import etree #la libreria lxml nos permite un mejor manejo de la información en html, 
import pandas as pd #Creación de data frames
```

```python
#Se realiza la solicitud de importación:
r=requests.get("https://listado.mercadolibre.com.pe/casas-lima") 
r 
```

```python
# Extracción de información en formato HTML con la libreria BeautifulSoup: 
soup=BeautifulSoup(r.text,"html.parser") 
# lectura con r.text ya que se tendrá el texto completo con tildes y ñs sin errores de ortografía
```

```python
soup
```

<p style="font-size:22px;color:green";>Extracción de títulos</p>

```python
#Con "find_all" buscamos la línea que contiene los títulos (obtenida de la previa inspección hecha en la página web)
titulos=soup.find_all("h2",attrs={"class":"ui-search-item__title ui-search-item__group__element"}) 
```

```python
#Limpieza: Se requiere sólo el texto de los títulos:
titulos=[i.text for i in titulos]
print(titulos)
```

<p style="font-size:22px;color:green";>Extracción del tipo de moneda</p>

```python
#Buscamos el tipo de moneda con la función find_all (damos la etiquetas obtenidas previa inspección en la página):
moneda=soup.find_all("span",attrs= {"class":"price-tag-symbol"})
```

```python
# Limpieza:
moneda=[i.text for i in moneda ]
print(moneda)
```

<p style="font-size:22px;color:green";>Extracción de los precios</p>

```python
#Buscamos los precios con la función find_all (damos la etiquetas obtenidas previa inspección en la página):
precios=soup.find_all("span",attrs= {"class":"price-tag-fraction"})
```

```python
# Limpieza:
precios=[i.text for i in precios]
print(precios)
```

<p style="font-size:22px;color:green";>Extracción del área en $m^{2}$</p>

```python
#Usamos el xpath para encontrar un elemento en particular de la página:
dom=etree.HTML(str(soup))
area=dom.xpath('//ul[@class="ui-search-card-attributes ui-search-item__group__element"]//li[1]')
```

```python
#Limpieza:
area=[i.text for i in area]
print(area)
# La extracción del área de las casas está en conjunto con su valor en $m^{2}$ más dos categorias (cubierta o total),
# decidimos dejarlo de esa manera.
```

<p style="font-size:22px;color:green";>Extracción de la ubicación / dirección</p>

```python
#Buscamos las direcciones de las casas mediante inspección a la página, luego usamos la función find_all: 
ubicacion=soup.find_all("span",attrs={"class":"ui-search-item__group__element ui-search-item__location"})
```

```python
#Limpieza
ubicacion=[i.text for i in ubicacion]
print(ubicacion)
```

<p style="font-size:22px;color:green";>Extracción del número de habitaciones</p>

```python
#Usamos el xpath:
habitaciones=dom.xpath('//ul[@class="ui-search-card-attributes ui-search-item__group__element"]//li[2]')
```

```python
#Limpieza:
habitaciones=[i.text for i in habitaciones]
print(habitaciones)
```

<p style="font-size:22px;color:green";>Extracción de URLs</p>

```python
#Se extrae de manera similar a los títulos, modificando etiquetas y atributos:
urls=soup.find_all("a",attrs={"class":"ui-search-result__content ui-search-link"})
```

```python
# Limpieza para los urls, se quiere obtener sólo los links contenidos en "href"
urls=[i.get("href") for i in urls]
print(urls)
```

```python
#Requerimos el mismo número de observaciones para lograr crear el data frame posterior
print(len(titulos))
print(len(moneda))
print(len(precios))
print(len(area))
print(len(ubicacion))
print(len(habitaciones))
print(len(urls))
# Notamos que el npumero de habitaciones es menor por lo cual decidimos no incluirla en el data frame.
```

<p style="font-size:22px;color:green";>Creación del primer data frame, sólo con la primera página de resultados</p>

df=pd.DataFrame({"Títulos":titulos,"Moneda":moneda ,"Precios":precios,"Área":area,"Ubicaciones":ubicacion,"URL":urls})

<p style="font-size:22px;color:green";>Extrayendo los datos de todas las páginas</p>

```python
#obteniendo el href de toda la pagina  ## https://listado.mercadolibre.com.pe/inmuebles/casas/venta/casas-lima_Desde_49_NoIndex_True
siguiente = dom.xpath('//div[@class="ui-search-pagination"]/ul/li[contains(@class,"--next")]/a')[0].get('href')  
siguiente 
```

```python
# Extrayendo la primera pagina #1 
ini=soup.find("span",attrs={"class":"andes-pagination__link"}).text
ini=int(ini)
ini
```

```python
#Extrayendo la ultima pagina #4 
can=soup.find("li",attrs={"class":"andes-pagination__page-count"})
can=int(can.text.split(" ")[1]) 
can   
```

```python
#creando una lista donde se guardaran todas las paginas de cada información como titulos,precios,areas,ubicacion y url's
lista_titulos = []
lista_precios = []
lista_areas =   []
lista_ubicacion = []
lista_url =   []
lista_soles_dolares = []
```

```python
#creamos el bucle para recorrer todas las paginas de mercado libre 
from requests.api import request
siguiente = 'https://listado.mercadolibre.com.pe/casas-lima'  #en principio lee el html de la pagina 1 del objeto'siguiente'
while True: 
    r= requests.get(siguiente)                  #con requests se obtendra el html de la pagina y asi sucesivamente recorrerá hasta la última pagina 
    if r.status_code==200:                       #validando el status a 200
        soup = BeautifulSoup(r.text,'html.parser')
        #####Titulos
        titulos=soup.find_all("h2",attrs={"class":"ui-search-item__title ui-search-item__group__element"}) 
        titulos=[i.text for i in titulos]
        lista_titulos.extend(titulos)              # extendiendo la lista sobre la informacion de los titulos

        #####precios
        precios=soup.find_all("span",attrs= {"class":"price-tag-fraction"})
        precios=[i.text for i in precios]
        lista_precios.extend(precios)             # extendiendo la lista sobre la informacion de los precios 


        #####areas
        dom=etree.HTML(str(soup))
        area=dom.xpath('//ul[@class="ui-search-card-attributes ui-search-item__group__element"]//li[1]')
        area=[i.text for i in area]
        lista_areas.extend(area)                  # extendiendo la lista sobre la informacion de las areas


        #####ubicacion
        ubicacion=soup.find_all("span",attrs={"class":"ui-search-item__group__element ui-search-item__location"})
        ubicacion=[i.text for i in ubicacion]
        lista_ubicacion.extend(ubicacion)        # extendiendo la lista sobre la informacion de las ubicaciones 


        ####url:
        urls=soup.find_all("a",attrs={"class":"ui-search-result__content ui-search-link"})
        urls=[i.get("href") for i in urls]
        lista_url.extend(urls)                  # extendiendo la lista sobre la informacion de los url

        ####soles_dolares:
        moneda=soup.find_all("span",attrs= {"class":"price-tag-symbol"})
        moneda=[i.text for i in moneda ]
        lista_soles_dolares.extend(moneda)

        ini=soup.find("span",attrs={"class":"andes-pagination__link"}).text
        ini=int(ini)                        #volver a validar la pagina inicial para el segundo recorrido del html en el bucle 

        can=soup.find("li",attrs={"class":"andes-pagination__page-count"})
        can=int(can.text.split(" ")[1])    #volver a validar la cantidad de paginas 

    


    else:
        break
    print(ini,can)                              

    if ini==can:  
        break             #break si ambos coinciden con la ultima pagina
    

    siguiente = dom.xpath('//div[@class="ui-search-pagination"]/ul/li[contains(@class,"--next")]/a')[0].get('href')
```

```python
#longitudes de las listas creadas (ambas deberian ser del mismo tamaño)
print(len(lista_titulos))
print(len(lista_precios)) 
print(len(lista_areas))
print(len(lista_ubicacion))
print(len(lista_url))
print(len(lista_soles_dolares))
```

Data frame de todas las paginas extraidas (1-4)

```python
df_final=pd.DataFrame({"Títulos":lista_titulos, "Monedas":lista_soles_dolares,"Precios":lista_precios,"Área":lista_areas,"URL":lista_url,"Ubicaciones":lista_ubicacion})
df_final
```

<p style="font-size:22px;color:green";>Creando un fichero excel</p>

```python
data=pd.DataFrame(df_final,columns=["Títulos","Monedas", "Precios","Área","URL","Ubicaciones"])
data
```
